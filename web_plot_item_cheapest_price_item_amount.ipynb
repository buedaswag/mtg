```python
from_ = df.xs(card_name, level='card_name', drop_level=True).\
        groupby(level='ts').item_price.first().to_frame()
stock_ = df.xs(card_name, level='card_name', drop_level=True).\
        groupby(level='ts').item_amount.sum().to_frame()
df_ = from_.join(stock_)
df_
```

```python
df_pickle = df_.copy().reset_index() 
df_pickle.ts = df_pickle.ts.dt.strftime('%Y-%m-%d %H-%M') 
list_pickle = df_pickle.values.tolist() 
list_pickle
```
